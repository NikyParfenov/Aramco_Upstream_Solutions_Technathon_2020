```python
import cv2
import os
import pandas as pd
import numpy as np
import scipy.io
from PIL import Image
from matplotlib import pyplot as plt
```

```python
data_dir = r'./data'
```

```python
os.listdir(data_dir)
```

```python
beta_mat = scipy.io.loadmat(os.path.join(data_dir, 'Beta_1_0.2.mat'))
print(beta_mat.keys())
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15,15))
ax1.imshow(beta_mat['ctImage'][:,:,0], cmap='Greys')
ax1.title.set_text("ct_Image z=0")

ax2.imshow(beta_mat['bwImage'][:,:,0], cmap='Greys')
ax2.title.set_text(f"bw_Image z=0")

ax3.imshow(beta_mat['ctImage'][:,:,10], cmap='Greys')
ax3.title.set_text(f"ct_Image z=10")

ax4.imshow(beta_mat['bwImage'][:,:,10], cmap='Greys')
ax4.title.set_text(f"bw_Image z=10")

plt.show()
```
